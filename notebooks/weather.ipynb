```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load data
data = pd.read_csv("../data/CF2_Weather_2020-2022.csv")

# Clean up datetime, drop 2nd datetime column
# Convert from POSIX to ISO UTC time
data.dt = pd.to_datetime(data.dt, unit='s', utc=True, infer_datetime_format=True)

# Set date as index
data = data.set_index('dt')

# Convert datetime to Eastern timezone
data.index = data.index.tz_convert('US/Eastern')

# Drop 2nd datetime column that's not needed
data = data.drop('dt_iso', axis=1)
```

```python
data.plot(figsize=(20, 10))
plt.show()
```

```python
sns.heatmap(data.corr(), cmap='viridis'
```
