```python
import numpy as np
import pandas as pd

# import data from a .csv into pandas
data = pd.read_csv('../data/Capital_Flats_2021-12-01_14_20_34.csv', header=[0], index_col=[0], low_memory=False, skipinitialspace=True) 

# loop through columns to get names right
new_columns = []
for column in data.columns:
    if 'Unnamed' not in column:
        spread = column
        new_columns.append(column)
    else:
        new_columns.append(spread)

# reset first column names
data.columns = new_columns
pd.MultiIndex.from_frame(data.iloc[0:1]) #TODO set the data frame to use MultiIndex columns 
data.head()
```
